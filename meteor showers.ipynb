FUNCTION FOR PREDICTION OF METEOR SHOWERS 

```python
def predict_best_meteor_shower_viewing(city):
    meteor_showers_string = ""
    
    if city not in cities.values:
        meteor_showers_string = "Unfortunately, " + city + " is not available for prediction at this time."
        return meteor_showers_string
    
    latitude = cities.loc[cities['city'] == city, 'latitude'].iloc[0]
    constellation_list = constellations.loc[(constellations['latitudestart'] >= latitude) & (constellations['latitudeend'] <= latitude), 'constellation'].tolist()
        
    if not constellation_list:
        meteor_showers_string = "Unfortunately, there are no meteor showers viewable from " + city + "."
        return meteor_showers_string
    
    for constellation in constellation_list:
        meteor_shower = meteor_showers.loc[meteor_showers['radiant'] == constellation, 'name'].iloc[0]
        meteor_shower_startdate = meteor_showers.loc[meteor_showers['radiant'] == constellation, 'startdate'].iloc[0]
        meteor_shower_enddate = meteor_showers.loc[meteor_showers['radiant'] == constellation, 'enddate'].iloc[0]
        moonphases_list = moon_phases.loc[(moon_phases['date'] >= meteor_shower_startdate) & (moon_phases['date'] <= meteor_shower_enddate)]
        
        if moonphases_list.empty:
            continue  # Skip this constellation if there are no moonphases available
        
        best_moondate = moonphases_list.loc[moonphases_list['percentage'].idxmin()]['date']
        meteor_showers_string += meteor_shower + " is best seen if you look towards the " + constellation + " constellation on " + best_moondate.to_pydatetime().strftime("%B %d, %Y") + ".\n"
    
    return meteor_showers_string

```

TESTING

```python
print (predict_best_meteor_shower_viewing("San Diago"))
```
